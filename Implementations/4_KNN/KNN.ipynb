### Now we need a function to split training and testing data

```python
import csv
import random
def loadDataSet(filename, split, trainingSet=[], testSet=[]):
    with open(filename, 'r') as csvfile:
        lines = csv.reader(csvfile)
        dataset = list(lines)
        for x in range(len(dataset)-1):
            for y in range(4):
                dataset[x][y]=float(dataset[x][y])
            if random.random() < split:
                trainingSet.append(dataset[x])
            else:
                testSet.append(dataset[x])
```

```python
trainingSet=[]
testSet=[]
loadDataSet(r'iris1.txt', 0.66 ,trainingSet, testSet)
print('Train: '+ repr(len(trainingSet)))
print('Test: '+ repr(len(testSet)))
```

### Now we have to make an Euclidean distance function and measure the distance

```python
import math
def euclideanDistance(instance1, instance2, length):
    distance = 0
    for x in range(length):
        distance += pow((instance1[x] - instance2[x]), 2)
    return math.sqrt(distance)
```

```python
data1 = [2,2,2, 'a']   #It belongs to class 'a'
data2 = [4,4,4, 'b']   #It belongs to class 'b'
distance = euclideanDistance(data1, data2, 3)
print('Distance: '+ repr(distance))
```

### Now we have the distance, let's look for K-nearest neighbor

```python
import operator
def getNeighbors(trainingSet,testInstance,k):
    distances = []
    length = len(testInstance) -1
    for x in range(len(trainingSet)):
        dist = euclideanDistance(testInstance, trainingSet[x], length)
        distances.append((trainingSet[x], dist))
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(k):
        neighbors.append(distances[x][0])
    return neighbors
```

```python
trainSet = [[2,2,2, 'a'], [4,4,4, 'b']]
testInstance = [5,5,5]
k = 1
neighbors = getNeighbors(trainSet, testInstance, 1)
print(neighbors)
```

```python
import operator
def getResponse(neighbors):
    classVotes = {}
    for x in range(len(neighbors)):
        response =  neighbors[x][-1]
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
    sortedVotes = sorted(classVotes.items(), key=operator.itemgetter(1), reverse=True)
    return sortedVotes[0][0]
```

```python
# we have to print that our testInstance[5,5,5] belongs to which class (either a or b) 
neighbors = [[1,1,1, 'a'], [2,2,2, 'a'], [3,3,3, 'b']]
print(getResponse(neighbors))
```

### Now we'll check the accuracy of our result

```python
def getAccuracy(testSet, predictions):
    correct = 0
    for x in range(len(testSet)):
        if testSet[x][-1] is predictions[x]:
            correct += 1
    return (correct/float(len(testSet))) * 100.0
```

```python
testSet = [[1,1,1, 'a'], [2,2,2, 'a'], [3,3,3, 'b']]
predictions = ['a', 'a', 'a']   #[true, true, false] predictions; 2/6 
# accuracy of the predictions on our test data
accuracy = getAccuracy(testSet, predictions)
print(accuracy)
```

### Combine all the KNN functions into one main function

```python
def main():
    #prepare data
    trainingSet = []
    testSet = []
    split = 0.67
    loadDataSet('iris1.txt', split ,trainingSet, testSet)
    print('Train set: '+ repr(len(trainingSet)))
    print('Test set: '+ repr(len(testSet)))
    #generate predictions
    predictions = []
    k=3
    for x in range(len(testSet)):
        neighbors = getNeighbors(trainingSet, testSet[x], k)
        result = getResponse(neighbors)
        predictions.append(result)
        print('> predicted = '+ repr(result) + ', actual = '+ repr(testSet[x][-1]))
    accuracy = getAccuracy(testSet, predictions)
    print('Accuracy: '+ repr(accuracy) + '%')
```

```python
main()
```
